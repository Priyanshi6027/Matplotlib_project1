plot & bar

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = [ 1,2,3,4,5,6,7,8,9]
temp = [ 2.3,4.5,8.7,1.4,3.3,2.3,6.6,7.9,7.0]
plt.title("temperature")
plt.xlabel("days")
plt.ylabel("temp")
plt.plot(days,temp, color = "y", marker = "o", linestyle = ":", linewidth = 3, markersize = 10)
plt.show()
```

```python
days = [ 1,2,3,4,5,6,7,8,9]
temp = [ 2.3,4.5,8.7,1.4,3.3,2.3,6.6,7.9,7.0]
plt.title("temperature", fontsize = 20)
plt.xlabel("days", fontsize = 15)
plt.ylabel("temp", fontsize = 15)
plt.plot(days,temp, "g^--", linewidth = 3, markersize = 10)
plt.legend(["temp line"], loc = 4)
plt.show()
```

```python
x =[ "python","java","js","php","r"]
y = [ 20,45,33,42,12]
plt.title("chart", fontsize = 15)
plt.xlabel("language", fontsize = 15)
plt.ylabel("number", fontsize = 15)
c = ["g","r","m","y","b"]
plt.bar(x,y, width = 0.4, color = c, edgecolor = "k", linewidth = 5)
plt.legend(["lan"])
plt.show()
```

```python
x =[ "python","java","js","php","r"]
y = [ 20,45,33,42,12]
z = [12,23,5,24,6]
plt.title("chart", fontsize = 15)
plt.xlabel("language", fontsize = 15)
plt.ylabel("number", fontsize = 15)
c = ["g","r","m","y","b"]
plt.bar(x,y, width = 0.4, color = "r" , label = "popularity1")
plt.bar(x,z, width = 0.4, color = "k" , label = "popularity1")
plt.legend()
plt.show()
```

```python
x =[ "python","java","js","php","r"]
y = [ 20,45,33,42,12]
z = [12,23,5,24,6]
plt.title("chart", fontsize = 15)
plt.xlabel("language", fontsize = 15)
plt.ylabel("number", fontsize = 15)
c = ["g","r","m","y","b"]
plt.barh(x,y, color = "y" , label = "popularity1")
plt.barh(x,z, color = "b" , label = "popularity1")
plt.legend()
plt.show()
```
